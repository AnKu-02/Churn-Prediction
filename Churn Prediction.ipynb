```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('C:/Users/anany/Downloads/Churn.csv')
```

```python
x = pd.get_dummies(df.drop(['Churn','Customer ID'], axis=1))
y = df['Churn'].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
#Import dependencies
import sys
!{sys.executable} -m pip install tensorflow

from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense  #Dense will be the fully connected layer
from sklearn.metrics import accuracy_score
```

```python
#Build and compile model
model = Sequential()
#input_dim has same dimension as x_train columns
model.add(Dense(units=32, activation='relu', input_dim=len(x_train.columns))) #32 neurons inside this dense layer
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
#Fit, predict and evaluate
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

```python
model.fit(x_train, y_train, epochs=200, batch_size=32)
```

```python
y_hat=model.predict(x_test)
y_hat=[0 if val<0.5 else 1 for val in y_hat]
y_hat
```

```python
accuracy_score(y_test, y_hat)
```

```python
#Saving and reloading
model.save('tfmodel')
```
